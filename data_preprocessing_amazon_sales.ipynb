#### 1.7 Convert dataset into atomic

The last major issue in the dataset is that review-related columns are not atomic.

Columns affected:
`about_product, user_id, user_name, review_id, review_title, review_content`

Current Problem:

- Multiple reviews for a single product are stored in the same row, separated by commas.

- This structure is non-atomic, making analysis, aggregation, and machine learning tasks difficult.

Solution:

- Split each comma-separated field so that each review gets its own row.

- Retain all product-related information (product_id, product_name, main_category, etc.) in each row.

- Ensure one row = one review, creating a fully atomic structure.

Benefits:

- Facilitates review-level analysis such as sentiment, rating trends, or reviewer behavior.

- Makes the dataset ready for machine learning, deep learning, and data analysis tasks.

- Maintains the link between each review and its corresponding product.

```python
import pandas as pd

# Review columns
review_cols = ['user_id','user_name','review_id','review_title','review_content']

# Other columns (sab original)
other_cols = [col for col in df.columns if col not in review_cols]

# Final rows list
final_rows = []

for idx, row in df.iterrows():
    # Split review columns
    split_lists = {col: row[col].split(',') for col in review_cols}
    
    # Maximum number of reviews in this row
    n = max(len(l) for l in split_lists.values())

    # Fill shorter lists with empty strings
    for col in review_cols:
        lst = split_lists[col]
        split_lists[col] = lst + ['']*(n - len(lst))

    # Add each review as a separate row, including all original columns
    for i in range(n):
        review_values = [split_lists[col][i].strip() for col in review_cols]
        other_values = [row[col] for col in other_cols]
        final_rows.append(review_values + other_values)

# Final column order
columns = review_cols + other_cols

# Create cleaned DataFrame
df_clean = pd.DataFrame(final_rows, columns=columns)

# Reset index
df_clean.reset_index(drop=True, inplace=True)

# Check
df_clean.head()
```

#### 2.4 Recommendation for Handling Missing Data

1 user_id, user_name, review_id

 - These columns are identifiers, and missing values account for roughly half of the dataset.

 - Dropping rows would lead to significant data loss, so instead we replace empty strings with a placeholder label: "Unknown".

 - This preserves the atomicity of the dataset and allows ML or NLP models to handle missing information effectively.

2 review_title

 - About 45% of the entries are missing titles.

 - Solution: Replace empty strings with "No Title". This ensures all rows remain usable without losing data.

3 review_content

 - Only 2% of entries are missing content, a small proportion.

 - Solution: Replace empty strings with "No Content". This prevents empty strings from interfering with text processing pipelines.

```python
review_cols = ['user_id', 'user_name', 'review_id', 'review_title', 'review_content']

# Replacement rules
df['user_id'] = df['user_id'].replace('', 'Unknown')
df['user_name'] = df['user_name'].replace('', 'Unknown')
df['review_id'] = df['review_id'].replace('', 'Unknown')
df['review_title'] = df['review_title'].replace('', 'No Title')
df['review_content'] = df['review_content'].replace('', 'No Content')
```

> ### Outcome

 - The dataset now contains no empty strings in review-related columns.

 - All rows are preserved, keeping the dataset fully atomic.

 - ML and NLP models can now interpret missing or unknown information without errors.

```python
df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].describe()
```

```python
for col in ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col} Outliers:\n", outliers)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 10))  # 3x2 grid
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(col)

# Hide empty subplot if any
if len(axes) > len(cols):
    for j in range(len(cols), len(axes)):
        axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```
